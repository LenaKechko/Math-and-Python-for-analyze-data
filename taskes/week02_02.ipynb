```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-(x) / 2.0)
```

```python
x = np.arange(start=1, stop=16, step=1)
x
```

```python
y = np.array(f(x))
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
x = np.array([1,15])
b = np.array(f(x))
A = np.array([[1]* len(x), x])
A = A.T
w = numpy.linalg.solve(A,b)
plt.plot(w, f(w))
plt.show()
```

```python
x = np.array([1,8,15])
b = np.array(f(x))
A = np.array([[1]* len(x), x , x**2])
A = A.T
w = numpy.linalg.solve(A,b)
plt.plot(w, f(w))
plt.show()
```

```python
x = np.array([1,4,10,15])
b = np.array(f(x))
A = np.array([[1]* len(x), x , x**2, x**3])
A = A.T
w = numpy.linalg.solve(A,b)
plt.plot(w, f(w))
plt.show()
with open("week02_02.txt", "w") as fp:
    for i in w:
        fp.write(str(i) + " ")
```
